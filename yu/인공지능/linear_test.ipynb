```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
hours_studied = np.array([1, 2, 3, 4, 5, 7, 8, 9, 10])
exam_scores = np.array([30, 45, 55, 60, 65, 75, 85, 90, 98])

# 데이터 시각화
plt.scatter(hours_studied, exam_scores)
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# 모델 생성 및 훈련
model = LinearRegression()
model.fit(hours_studied.reshape(-1, 1), exam_scores)

# 예측선 그리기
predicted_scores = model.predict(hours_studied.reshape(-1, 1))

# 결과 시각화
plt.scatter(hours_studied, exam_scores)
plt.plot(hours_studied, predicted_scores, color='red')
plt.title("Linear Regression Fit")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
# 전체 코드 (복습 및 실행 편의를 위해)
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. 데이터 준비
hours_studied = np.array([1, 2, 3, 4, 5, 7, 8, 9, 10])
exam_scores = np.array([30, 45, 55, 60, 65, 75, 85, 90, 98])

# 2. 모델 훈련
model = LinearRegression()
model.fit(hours_studied.reshape(-1, 1), exam_scores)

# --- 여기서부터 새로운 코드 ---

# 3. 모델 성능 평가 (R-squared 점수 계산)
r_squared = model.score(hours_studied.reshape(-1, 1), exam_scores)

print(f"모델의 R-제곱 점수는 {r_squared:.4f} 입니다.")
```
